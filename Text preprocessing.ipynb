```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Dataset.csv')
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df
```

```python
df['comb'] = df['headline']+' '+df['short_description']
```

```python
import string 
def remove_punctuation(text):
    punctuationfree="".join([i for i in str(text) if i not in string.punctuation and i not in "'"])
    return punctuationfree
df['comb']= df['comb'].apply(lambda x:remove_punctuation(x))
df.head()
```

```python
df['comb'] = df['comb'].apply(lambda x:str(x).lower())
```

```python
df
```

```python
df.drop(columns=['headline','short_description'],inplace=True)
```

```python
from nltk.tokenize import word_tokenize 
```

```python
df['comb'] = df['comb'].apply(lambda x: word_tokenize(str(x)))
```

```python
df
```

```python
import nltk

stopwords = nltk.corpus.stopwords.words('english')

def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
df['comb']= df['comb'].apply(lambda x:remove_stopwords(x))
```

```python
df
```

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
```

```python
def lemmatizer(text):
    l = [wnl.lemmatize(word,pos="v") for word in text]
    return l
```

```python
df['comb'] = df['comb'].apply(lambda x:lemmatizer(x))
```

```python
df
```

```python
df.to_csv("Preprocessed_Dataset.csv",encoding='utf-8')
```
