```python
import pandas as pd
```

```python
import json

data = []

with open('./News_Category_Dataset_v2.json') as f:
    for x in f:
        data.append(json.loads(x))
df = pd.DataFrame(data)
```

```python
df.head(10)
```

```python
len(df)
```

```python
couple = df['category'].value_counts()
couple.sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
sns.countplot(data=df,x='category')
plt.xticks(rotation=90)
```

```python
filtered_columns = ['POLITICS','WELLNESS','ENTERTAINMENT','TRAVEL','STYLE & BEAUTY']

filtered_df = df[df['category'].isin(filtered_columns)]
```

```python
filtered_df = pd.DataFrame(filtered_df,columns=['headline','short_description','category'])
```
